```python
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
# Линейно конгруэнтный метод
class RandomGenerator:
    def __init__(self, seed):
        self.seed = seed

    def random(self):
        a = 1664525
        c = 1013904223
        m = 2 ** 32
        self.seed = (a * self.seed + c) % m
        self.seed ^= self.seed >> 7
        return (self.seed) % 100
```

```python
seed_value = 12345 # Начальное значение для генератора
generator = RandomGenerator(seed_value)
datas1 = list()
datas2 = list()
# datas3 = list()
```

```python
# Метод побитового сдвига
def generate_random_number(seed):
    seed ^= (seed << 21)
    seed ^= (seed >> 35)
    seed ^= (seed << 4)
    return seed % 100
```

```python
seed = 777
for i in range(20):
    lst = list()
    for j in range(10000):
        random_num = generate_random_number(seed)
        seed += random.randint(1, 1000)
        lst.append(random_num)
    datas2.append(lst)
```

```python
# Среднее
def avg(data):
    s = sum(data)
    n = len(data)
    return s / n
```

```python
# Отклонение
def st_dev(data):
    s = sum(data)
    n = len(data)
    avg = s / n
    res = 0
    for i in data:
        res += (i - avg)**2
    return (res / n)**0.5
```

```python
# Коэф. вариации
def var(data):
    return st_dev(data) / avg(data) * 100
```

```python
for data in datas1:
    print(avg(data), st_dev(data), var(data))
```

```python
def make_graph(data, filename=None, color='red'):
    fig = plt.figure()
    ax1 = fig.add_subplot()
    ax1.set_xlabel('Размер выборки')
    ax1.set_ylabel('Время(сек.)')

    x = [str(point[0]) for point in data]
    y = [point[1] for point in data]
    plt.plot(x, y, marker=".", color=color)
```
